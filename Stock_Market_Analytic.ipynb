```python
import yfinance as yf
import pandas as pd

df = yf.download("AAPL", start="2022-01-01", end="2025-01-01")

df.head()
```

```python
import plotly.express as px

fig = px.line(
    x=df.index,
    y=df["Close"].squeeze(),
    title="Apple Stock Closing Price"
)

fig.show()
```

```python
df["MA50"] = df["Close"].rolling(50).mean()

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=df.index, y=df["Close"].squeeze(), name="Close"))
fig.add_trace(go.Scatter(x=df.index, y=df["MA50"].squeeze(), name="50-Day MA"))

fig.update_layout(title="Apple Stock Price with 50-Day Moving Average")

fig.show()
```

```python
df["Close"] = df["Close"].squeeze()

df["MA50"] = df["Close"].rolling(window=50).mean()

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=df.index, y=df["Close"], name="Close"))
fig.add_trace(go.Scatter(x=df.index, y=df["MA50"], name="MA50"))

fig.show()
```

```python
import plotly.express as px

fig = px.bar(
    x=df.index,
    y=df["Volume"].squeeze(),
    title="Trading Volume"
)

fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(
    x=df.index,
    open=df["Open"].squeeze(),
    high=df["High"].squeeze(),
    low=df["Low"].squeeze(),
    close=df["Close"].squeeze()
)])

fig.update_layout(title="Apple Candlestick Chart")

fig.show()
```

```python
delta = df["Close"].diff()

gain = delta.where(delta > 0, 0).rolling(14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(14).mean()

rs = gain / loss
df["RSI"] = 100 - (100 / (1 + rs))

import plotly.express as px

fig = px.line(x=df.index, y=df["RSI"], title="RSI (14 Days)")
fig.show()
```

```python
exp1 = df["Close"].ewm(span=12, adjust=False).mean()
exp2 = df["Close"].ewm(span=26, adjust=False).mean()

df["MACD"] = exp1 - exp2
df["Signal"] = df["MACD"].ewm(span=9, adjust=False).mean()

import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=df.index, y=df["MACD"], name="MACD"))
fig.add_trace(go.Scatter(x=df.index, y=df["Signal"], name="Signal"))

fig.update_layout(title="MACD Indicator")
fig.show()
```
